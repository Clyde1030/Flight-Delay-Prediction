```python
from pyspark.ml.feature import VectorAssembler

feature_cols = [
    "HourlyPrecipitation","HourlySeaLevelPressure","HourlyAltimeterSetting",
    "HourlyWetBulbTemperature","HourlyStationPressure","HourlyWindSpeed",
    "HourlyRelativeHumidity","HourlyDewPointTemperature","HourlyDryBulbTemperature",
    "HourlyVisibility","QUARTER","day_of_week","DAY_OF_MONTH", "CRS_DEP_TIME",
    "OP_CARRIER_FL_NUM","dest_airport_lat","dest_airport_lon",
    "origin_airport_lat","origin_airport_lon","DISTANCE","DISTANCE_GROUP",
    "month","is_us_holiday","is_holiday_window","is_summer_peak",
    "hour_of_day","is_weekend","wind_dir_sin","wind_dir_cos",
    # OHE vectors:
    "ORIGIN_ohe","TAIL_NUM_ohe","DEP_TIME_BLK_ohe","dest_region_ohe",
    "dest_type_ohe","NAME_ohe","origin_region_ohe","origin_type_ohe",
    "origin_iata_code_ohe","dest_iata_code_ohe"
]

assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features",
    handleInvalid="skip"
)


df_ml = assembler.transform(df_otpw).select("features", "DEP_DELAY_NEW")
```

```python
df_clean = df_ml.na.drop()
```

```python
from pyspark.ml.regression import LinearRegression


lasso = LinearRegression(
    featuresCol="features",
    labelCol="DEP_DELAY_NEW",
    elasticNetParam=1.0,      # full L1 regularization
    regParam=0.1              # tune this
)

lasso_model = lasso.fit(df_clean)
```

```python
# Get all coefficients
coefficients = lasso_model.coefficients
intercept = lasso_model.intercept

print("Intercept:", intercept)
print("Coefficients:", coefficients)
```

```python
from pyspark.sql.functions import col, year, to_date
from pyspark.ml.functions import vector_to_array

# ------------------------------
# 1. Convert FL_DATE to proper date format
# ------------------------------
df_otpw = df_otpw.withColumn("FL_DATE", to_date(col("FL_DATE"), "yyyy-MM-dd"))

# ------------------------------
# 2. Extract YEAR and filter for 2015
# ------------------------------
df_2015 = df_otpw.withColumn("YEAR", year(col("FL_DATE"))).filter(col("YEAR") == 2015)

# ------------------------------
# 3. Define numeric and vector features
# ------------------------------
numeric_features = [
    "HourlyPrecipitation",
    "is_summer_peak",
    "HourlyAltimeterSetting",
    "is_us_holiday",
    "is_holiday_window",
    "wind_dir_sin",
    "QUARTER",
    "HourlyVisibility",
    "HourlyWindSpeed",
    "hour_of_day",
    "origin_airport_lat",
    "HourlyRelativeHumidity",
    "day_of_week",
    "HourlyWetBulbTemperature",
    "HourlyDryBulbTemperature",
    "dest_airport_lat",
    "dest_airport_lon",
    "is_weekend",
    "CRS_DEP_TIME",
    "DAY_OF_MONTH",
    "DISTANCE",
    "OP_CARRIER_FL_NUM"
]

vector_features = [
    "dest_iata_code_ohe",
    "ORIGIN_ohe",
    "origin_type_ohe",
    "TAIL_NUM_ohe",
    "NAME_ohe",
    "dest_type_ohe",
    "dest_region_ohe",
    "origin_iata_code_ohe"
]

label_col = "DEP_DELAY_GROUP"

# ------------------------------
# 4. Keep only existing columns to avoid unresolved column errors
# ------------------------------
existing_numeric = [c for c in numeric_features if c in df_2015.columns]
existing_vector  = [c for c in vector_features if c in df_2015.columns]

missing_cols = (set(numeric_features + vector_features) - set(existing_numeric + existing_vector))
if missing_cols:
    print(" Warning: These columns are missing and were removed:")
    print(missing_cols)

# Ensure label exists
all_cols = existing_numeric + existing_vector + [label_col]
df_2015 = df_2015.select(*all_cols)

# ------------------------------
# 5. Convert vector columns from struct to array if needed
# ------------------------------
from pyspark.sql.types import StructType

for col_name in existing_vector:
    if isinstance(df_2015.schema[col_name].dataType, StructType):
        df_2015 = df_2015.withColumn(col_name, vector_to_array(col(col_name)))

# ------------------------------
# 6. Split train and test by QUARTER
# ------------------------------
train_df = df_2015.filter(col("QUARTER").isin([1, 2, 3]))
test_df  = df_2015.filter(col("QUARTER") == 4)

print("Train rows:", train_df.count())
print("Test rows:", test_df.count())
print("Training columns:", train_df.columns)

```

```python
from pyspark.ml.feature import VectorAssembler

# 1. Drop NA rows first
numeric_features = [f for f in feature_cols if not f.endswith("_ohe")]
vector_features  = [f for f in feature_cols if f.endswith("_ohe")]

train_df_clean = train_df.dropna(subset=numeric_features + vector_features + [label_col])
test_df_clean  = test_df.dropna(subset=numeric_features + vector_features + [label_col])

# 2. Sample small subset (1k-10k) for faster testing
train_df_sample = train_df_clean.limit(1000)
test_df_sample  = test_df_clean.limit(1000)

# 3. VectorAssembler on all features at once
assembler = VectorAssembler(
    inputCols=numeric_features + vector_features,
    outputCol="features_xgb_v4"
)

train_df_sample = assembler.transform(train_df_sample)
test_df_sample  = assembler.transform(test_df_sample)

print("Train sample rows:", train_df_sample.count())
print("Test sample rows:", test_df_sample.count())
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from xgboost.spark import SparkXGBRegressor

xgb = SparkXGBRegressor(
    features_col="features_xgb",
    label_col="DEP_DELAY_GROUP",
    objective="reg:squarederror",
    n_estimators=500,
    max_depth=6,
    eta=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    missing=0.0,
    seed=42
    verbosity=1
)

paramGrid = (
    ParamGridBuilder()
    .addGrid(xgb.max_depth, [4, 6, 8])
    .addGrid(xgb.n_estimators, [100, 300, 500])
    .addGrid(xgb.learning_rate, [0.01, 0.05, 0.1])
    .addGrid(xgb.subsample, [0.7, 0.8, 1.0])
    .addGrid(xgb.colsample_bytree, [0.7, 0.8, 1.0])
    .build()
)

evaluator = RegressionEvaluator(
    labelCol="DEP_DELAY_GROUP",
    predictionCol="prediction",
    metricName="rmse"
)

cv = CrossValidator(
    estimator=xgb,
    estimatorParamMaps=paramGrid,
    evaluator=evaluator,
    numFolds=3,       # 3-fold CV
    parallelism=4     # adjust depending on cluster size
)

cv_model = cv.fit(train_df)
```

```python
preds = cv_model.transform(test_df)

rmse = evaluator.evaluate(preds)
r2 = RegressionEvaluator(
    labelCol="DEP_DELAY_NEW",
    predictionCol="prediction",
    metricName="r2"
).evaluate(preds)

print(f"Test RMSE: {rmse}, R: {r2}")
```

```python
best_model = cv_model.bestModel
feature_importance = best_model.nativeBooster.getScore(importance_type='weight')
sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)
```
