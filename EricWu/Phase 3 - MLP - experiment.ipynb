```python
# Function to check nulls for a dataframe
from pyspark.sql.functions import col, sum as _sum, when
def null_summary(df):
    total_rows = df.count()

    # Compute null count and percentage per column
    null_summary = df.select([
        _sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
        for c in df.columns
    ]).toPandas().T.reset_index()

    null_summary.columns = ["Column", "Null_Count"] 
    null_summary["Null_%"] = (null_summary["Null_Count"] / total_rows * 100).round(2)

    # Sort by highest missing percentage
    null_summary = null_summary.sort_values(by="Null_%", ascending=False).reset_index(drop=True)

    return null_summary
```

```python
def make_undersampled_train(
    df,
    label_col="DEP_DELAY_GROUP_CLASSIFIED",
    minority_label="delayed",
    majority_label="not_delayed",
    maj_to_min_ratio=2.0,      # e.g. 2x more majority than minority
    day_col="DAY_OF_MONTH",    # stratify by this column
    seed=42
):

    # Split into minority and majority subsets
    df_min = df.filter(col(label_col) == minority_label)
    df_maj = df.filter(col(label_col) == majority_label)

    # Global counts (for logging)
    n_min = df_min.count()
    n_maj = df_maj.count()
    print(f"Original train counts: {minority_label}={n_min:,}, {majority_label}={n_maj:,}")

    # --- per-day counts for minority and majority ---
    min_per_day = (
        df_min.groupBy(day_col)
              .agg(F.count("*").alias("n_min"))
    )
    maj_per_day = (
        df_maj.groupBy(day_col)
              .agg(F.count("*").alias("n_maj"))
    )

    # Join majority counts with minority counts (left join: keep all majority days)
    per_day = (
        maj_per_day
        .join(min_per_day, on=day_col, how="left")
        .fillna({"n_min": 0})  # days with no minority get n_min = 0
    )

    # Compute target majority size and sampling fraction *per day*
    # target_maj(d) = min(n_maj(d), maj_to_min_ratio * n_min(d))
    # if n_min(d) == 0, target_maj(d) = 0   fraction = 0
    per_day = (
        per_day
        .withColumn(
            "target_maj",
            F.when(
                F.col("n_min") > 0,
                F.least(
                    F.col("n_maj"),
                    F.col("n_min") * F.lit(float(maj_to_min_ratio))
                )
            ).otherwise(F.lit(0.0))
        )
        .withColumn(
            "frac",
            F.when(F.col("n_maj") > 0,
                   F.col("target_maj") / F.col("n_maj")
            ).otherwise(F.lit(0.0))
        )
    )

    # Collect fractions into a dict for sampleBy
    fractions = {
        row[day_col]: float(row["frac"])
        for row in per_day.select(day_col, "frac").collect()
    }

    # Some logging
    print("Per-day undersampling fractions (day -> frac):")
    # You can comment this if too verbose
    for d, f in sorted(fractions.items()):
        print(f"  {day_col}={d}: frac={f:.4f}")

    # Stratified sampling of majority class by day_col
    df_maj_sample = df_maj.sampleBy(
        col=day_col,
        fractions=fractions,
        seed=seed
    )

    # Union minority + undersampled majority
    balanced = df_min.unionByName(df_maj_sample)

    # Optional: cache and print final counts
    balanced = balanced.cache()
    final_min = balanced.filter(col(label_col) == minority_label).count()
    final_maj = balanced.filter(col(label_col) == majority_label).count()

    print(f"Balanced train counts (stratified by {day_col}): "
          f"{minority_label}={final_min:,}, {majority_label}={final_maj:,}")

    return balanced
```

```python
def build_mlp_feature_pipeline(label_col, numeric_cols, categorical_cols):
    """
    Build *feature* pipeline for the MLP:
      - StringIndex label
      - StringIndex + OneHotEncode categorical_cols
      - Assemble numeric_cols -> 'num_vec'
      - StandardScaler on 'num_vec' -> 'num_scaled'
      - Assemble [num_scaled] + cat_oh -> final 'features'
    """
    # Label indexer (binary: not_delayed / delayed)
    label_indexer = StringIndexer(
        inputCol=label_col,
        outputCol="label",
        handleInvalid="skip"   # skip weird / unseen labels
    )

    # Categorical: season, etc.
    cat_indexers = [
        StringIndexer(inputCol=c, outputCol=f"{c}__idx", handleInvalid="keep")
        for c in categorical_cols
    ]

    ohe = OneHotEncoder(
        inputCols=[f"{c}__idx" for c in categorical_cols],
        outputCols=[f"{c}__oh"  for c in categorical_cols],
        handleInvalid="keep"
    )

    # Assemble raw numeric features
    num_assembler = VectorAssembler(
        inputCols=numeric_cols,
        outputCol="num_vec",
        handleInvalid="keep"
    )

    # Scale numeric features
    scaler = StandardScaler(
        inputCol="num_vec",
        outputCol="num_scaled",
        withMean=True,
        withStd=True
    )

    # Final feature vector: scaled numeric + one-hot categorical
    final_feature_cols = ["num_scaled"] + [f"{c}__oh" for c in categorical_cols]

    final_assembler = VectorAssembler(
        inputCols=final_feature_cols,
        outputCol="features",
        handleInvalid="keep"
    )

    feature_pipeline = Pipeline(
        stages=[label_indexer] + cat_indexers + [ohe, num_assembler, scaler, final_assembler]
    )
    return feature_pipeline
```

```python
def fit_and_eval_mlp(
    train_df,
    test_df,
    label_col,
    numeric_cols,
    categorical_cols,
    hidden_layers,          # e.g. [32] or [64, 32]
    maxIter=50,
    blockSize=128,
    seed=42,
):
    """
    Train + evaluate an MLP on top of a *single-fitted* feature pipeline.

    Returns:
      - feat_model   : fitted feature pipeline model
      - mlp_model    : trained MultilayerPerceptronClassificationModel
      - pred_test    : predictions on test_df (with 'prediction', 'probability', 'label')
      - metrics      : dict with train/test precision/recall/F1 for 'delayed'
    """

    # -------------------------------------------------
    # 1) Fit feature pipeline once on TRAIN
    # -------------------------------------------------
    feat_pipe = build_mlp_feature_pipeline(
        label_col=label_col,
        numeric_cols=numeric_cols,
        categorical_cols=categorical_cols,
    )

    feat_model = feat_pipe.fit(train_df)

    # Transform train & test to add 'label' + 'features'
    train_feat = feat_model.transform(train_df)

    # *** Optimization: cache the feature-engineered TRAIN set ***
    train_feat = train_feat.select("features", "label").cache()
    _ = train_feat.count()   # materialize cache


    test_feat  = feat_model.transform(test_df)

    # -------------------------------------------------
    # 2) Detect input_dim from 'features' metadata
    # -------------------------------------------------
    sample = (
        train_feat
        .select("features")
        .limit(1)
        .collect()[0]["features"]
    )
    input_dim = sample.size
    print(f"Detected input_dim for MLP = {input_dim}")

    # Layers: [input_dim] + hidden + [2 classes]
    layers = [input_dim] + hidden_layers + [2]

    # -------------------------------------------------
    # 3) Train the MLP
    # -------------------------------------------------
    trainer = MultilayerPerceptronClassifier(
        featuresCol="features",
        labelCol="label",
        predictionCol="prediction",
        probabilityCol="probability",
        maxIter=maxIter,
        blockSize=blockSize,
        layers=layers,
        seed=seed,
    )

    print(f"Training MLP with layers = {layers}")
    mlp_model = trainer.fit(train_feat)

    # -------------------------------------------------
    # 4) Evaluate on TRAIN & TEST: precision/recall/F1 for 'delayed'
    # -------------------------------------------------
    # IMPORTANT: transform the *feature* DataFrames, not raw dfs
    pred_train = mlp_model.transform(train_feat)
    pred_test  = mlp_model.transform(test_feat)

    # 'delayed' index from the label indexer (stage 0 of feature pipeline)
    label_names = feat_model.stages[0].labels  # e.g. ['not_delayed', 'delayed']
    delayed_idx = float(label_names.index("delayed"))

    def prf_for_delayed(pred_df):
        """Compute precision/recall/F1 for the 'delayed' class."""
        rdd = pred_df.select("prediction", "label") \
                     .rdd.map(lambda r: (float(r[0]), float(r[1])))
        mc = MulticlassMetrics(rdd)
        p = mc.precision(delayed_idx)
        r = mc.recall(delayed_idx)
        f1 = 2 * p * r / (p + r) if (p + r) > 0 else 0.0
        return p, r, f1

    # Metrics on train and test
    train_p, train_r, train_f1 = prf_for_delayed(pred_train)
    test_p,  test_r,  test_f1  = prf_for_delayed(pred_test)

    print("\n=== Metrics for 'delayed' ===")
    print(f"TRAIN: precision={train_p:.4f}, recall={train_r:.4f}, F1={train_f1:.4f}")
    print(f"TEST : precision={test_p:.4f},  recall={test_r:.4f},  F1={test_f1:.4f}")

    metrics = {
        "layers": layers,
        "maxIter": maxIter,
        "blockSize": blockSize,
        "precision_delayed_train": train_p,
        "recall_delayed_train": train_r,
        "f1_delayed_train": train_f1,
        "precision_delayed_test": test_p,
        "recall_delayed_test": test_r,
        "f1_delayed_test": test_f1,
    }
    # Optional: free memory when done
    train_feat.unpersist

    return feat_model, mlp_model, pred_test, metrics
```

## Building MLP model - Fine tuning with grid search

```python
def grid_search_mlp_with_thresholds(
    train_df,
    test_df,
    label_col,
    numeric_cols,
    categorical_cols,
    hidden_layer_configs,   # e.g. [[32], [64], [64, 32]]
    maxIter_list,           # e.g. [50]
    blockSize_list,         # e.g. [128]
    threshold_list,         # e.g. [0.3, 0.4, 0.5]
    seed=42
):
    """
    Grid search over MLP hyperparameters *and* decision thresholds.

    For each (hidden_layers, maxIter, blockSize):
      - fit feature pipeline once
      - fit one MLP model
      - for each threshold  in threshold_list:
          * compute precision/recall/F1 for 'delayed' on TRAIN and TEST
    Selects the config with best recall_delayed_test.
    """

    results = []
    best_cfg = None
    best_recall = -1.0

    for hidden_layers, maxIter, blockSize in product(hidden_layer_configs,
                                                     maxIter_list,
                                                     blockSize_list):

        # -------------------------------------------------
        # 1) Feature pipeline: label + OHE + scaling + features
        # -------------------------------------------------
        feat_pipe = build_mlp_feature_pipeline(
            label_col=label_col,
            numeric_cols=numeric_cols,
            categorical_cols=categorical_cols
        )
        feat_model = feat_pipe.fit(train_df)

        train_feat = feat_model.transform(train_df)
        test_feat  = feat_model.transform(test_df)

        # Detect input dimension from features
        sample = train_feat.select("features").limit(1).collect()[0]["features"]
        input_dim = sample.size
        layers = [input_dim] + hidden_layers + [2]

        # -------------------------------------------------
        # 2) Train MLP (cache features for speed)
        # -------------------------------------------------
        trainer = MultilayerPerceptronClassifier(
            featuresCol="features",
            labelCol="label",
            predictionCol="prediction",
            probabilityCol="probability",
            maxIter=maxIter,
            blockSize=blockSize,
            layers=layers,
            seed=seed
        )

        print(f"\n=== Training MLP: layers={layers}, maxIter={maxIter}, blockSize={blockSize} ===")

        train_feat_cached = train_feat.select("features", "label").cache()
        _ = train_feat_cached.count()  # materialize cache
        mlp_model = trainer.fit(train_feat_cached)
        train_feat_cached.unpersist()

        # Full predictions (we'll re-threshold manually)
        pred_train_full = mlp_model.transform(train_feat)
        pred_test_full  = mlp_model.transform(test_feat)

        # -------------------------------------------------
        # 3) Helper: compute metrics for a given threshold
        # -------------------------------------------------
        label_names = feat_model.stages[0].labels  # e.g. ['not_delayed', 'delayed']
        if "delayed" not in label_names:
            raise ValueError(f"'delayed' not found in labels: {label_names}")
        delayed_idx = float(label_names.index("delayed"))

        # assume binary: 2 classes
        if len(label_names) != 2:
            raise ValueError(f"Expected 2 classes, got {len(label_names)}: {label_names}")
        other_idx = float(1 - delayed_idx)  # if delayed is 1, other is 0; if delayed is 0, other is 1

        def prf_for_delayed_with_threshold(pred_df, threshold):
            """
            Apply custom threshold on P(delayed) and compute P/R/F1 for 'delayed'.
            """
            delayed_pos = int(delayed_idx)
            other_pos   = int(other_idx)

            # 1) Convert vector -> array, then extract delayed prob
            df2 = pred_df.withColumn(
                "prob_delayed",
                vector_to_array("probability")[delayed_pos]
            )

            # 2) Apply threshold on prob_delayed to create a new prediction column
            df_thr = df2.withColumn(
                "pred_thr",
                F.when(
                    F.col("prob_delayed") >= F.lit(threshold),
                    F.lit(float(delayed_pos))
                ).otherwise(F.lit(float(other_pos)))
            )

            # 3) Evaluate metrics using MulticlassMetrics on (pred_thr, label)
            rdd = df_thr.select("pred_thr", "label") \
                        .rdd.map(lambda r: (float(r[0]), float(r[1])))

            mc = MulticlassMetrics(rdd)

            p = mc.precision(delayed_idx)
            r = mc.recall(delayed_idx)
            f1 = 2 * p * r / (p + r) if (p + r) > 0 else 0.0
            return p, r, f1

        # -------------------------------------------------
        # 4) Evaluate all thresholds for this trained model
        # -------------------------------------------------
        for thr in threshold_list:
            train_p, train_r, train_f1 = prf_for_delayed_with_threshold(pred_train_full, thr)
            test_p,  test_r,  test_f1  = prf_for_delayed_with_threshold(pred_test_full, thr)

            print(
                f"  threshold={thr:.2f}  "
                f"TRAIN: P={train_p:.4f}, R={train_r:.4f}, F1={train_f1:.4f} | "
                f"TEST: P={test_p:.4f}, R={test_r:.4f}, F1={test_f1:.4f}"
            )

            cfg_result = {
                "layers": layers,
                "hidden_layers": hidden_layers,
                "maxIter": maxIter,
                "blockSize": blockSize,
                "threshold": thr,
                "precision_delayed_train": train_p,
                "recall_delayed_train": train_r,
                "f1_delayed_train": train_f1,
                "precision_delayed_test": test_p,
                "recall_delayed_test": test_r,
                "f1_delayed_test": test_f1,
            }
            results.append(cfg_result)

            # choose best by TEST recall for 'delayed'
            if test_r > best_recall:
                best_recall = test_r
                best_cfg = cfg_result

    print("\n===== BEST MLP CONFIG (by recall for 'delayed' on TEST) =====")
    print(best_cfg)

    return results, best_cfg
```

## Building MLP model - Adding time-based CV + thresholds 

```python
def prf_for_delayed_with_threshold(pred_df, delayed_idx, other_idx, threshold):
    """
    Given a prediction DF with columns:
      - label  (indexed)
      - probability (VectorUDT)
    and numeric label indices for 'delayed' and 'other', compute P/R/F1 for 'delayed'
    using a custom probability threshold.
    """
    delayed_pos = int(delayed_idx)
    other_pos   = int(other_idx)

    # 1) Convert probability vector -> array, then extract P(delayed)
    df2 = pred_df.withColumn(
        "prob_delayed",
        vector_to_array("probability")[delayed_pos]
    )

    # 2) Threshold into a custom prediction column
    df_thr = df2.withColumn(
        "pred_thr",
        F.when(F.col("prob_delayed") >= F.lit(threshold),
               F.lit(float(delayed_pos))
        ).otherwise(F.lit(float(other_pos)))
    )

    # 3) Compute metrics on (pred_thr, label)
    rdd = df_thr.select("pred_thr", "label") \
                .rdd.map(lambda r: (float(r[0]), float(r[1])))

    mc = MulticlassMetrics(rdd)
    p = mc.precision(delayed_idx)
    r = mc.recall(delayed_idx)
    f1 = 2 * p * r / (p + r) if (p + r) > 0 else 0.0

    return p, r, f1
```

```python
def time_cv_mlp_with_thresholds(
    train_df,
    quarter_col,
    label_col,
    numeric_cols,
    categorical_cols,
    hidden_layer_configs,   # e.g. [[32], [64], [64, 32]]
    maxIter_list,           # e.g. [50]
    blockSize_list,         # e.g. [128]
    threshold_list,         # e.g. [0.3, 0.4, 0.5]
    seed=42
):
    """
    Time-aware CV for MLP with thresholds:
      Fold 1: train Q1,       val Q2
      Fold 2: train Q1+Q2,    val Q3

    Returns:
      results  : list[dict] with mean train/val metrics for each config+threshold
      best_cfg : dict with best config by mean val recall for 'delayed'
      results_df : pandas DataFrame of results
    """

    # --- define folds on the given train_df (must contain Q1Q3) ---
    folds = [
        (
            train_df.filter(F.col(quarter_col) == 1),                   # train
            train_df.filter(F.col(quarter_col) == 2),                   # val
        ),
        (
            train_df.filter(F.col(quarter_col).isin(1, 2)),             # train
            train_df.filter(F.col(quarter_col) == 3),                   # val
        ),
    ]

    results = []
    best_cfg = None
    best_recall = -1.0

    # --- grid over architecture / training hyperparams ---
    for hidden_layers, maxIter, blockSize in product(
        hidden_layer_configs, maxIter_list, blockSize_list
    ):
        print(f"\n=== MLP config: hidden={hidden_layers}, "
              f"maxIter={maxIter}, blockSize={blockSize} ===")

        # store fold-wise metrics per threshold
        metrics_acc = {
            thr: {
                "train_p": [], "train_r": [], "train_f1": [],
                "val_p":   [], "val_r":   [], "val_f1":   [],
            }
            for thr in threshold_list
        }

        for fold_idx, (fold_train_raw, fold_val_raw) in enumerate(folds, start=1):
            n_tr = fold_train_raw.count()
            n_va = fold_val_raw.count()
            print(f"  Fold {fold_idx}: train n={n_tr}, val n={n_va}")

            # --- (1) Fit feature pipeline on this fold's TRAIN only ---
            feat_pipe = build_mlp_feature_pipeline(
                label_col=label_col,
                numeric_cols=numeric_cols,
                categorical_cols=categorical_cols,
            )
            feat_model = feat_pipe.fit(fold_train_raw)

            # Transform into (label, features)
            fold_train_feat = feat_model.transform(fold_train_raw) \
                                         .select("label", "features")
            fold_val_feat   = feat_model.transform(fold_val_raw) \
                                         .select("label", "features")

            # Cache train features so MLP can iterate faster
            fold_train_feat = fold_train_feat.cache()
            _ = fold_train_feat.count()  # materialize

            # Determine input_dim
            sample = fold_train_feat.select("features").limit(1).collect()[0]["features"]
            input_dim = sample.size
            layers = [input_dim] + hidden_layers + [2]
            print(f"    Layers = {layers}")

            # --- (2) Train MLP on this fold ---
            trainer = MultilayerPerceptronClassifier(
                featuresCol="features",
                labelCol="label",
                predictionCol="prediction",
                probabilityCol="probability",
                maxIter=maxIter,
                blockSize=blockSize,
                layers=layers,
                seed=seed,
            )

            print("    Training MLP on this fold...")
            mlp_model = trainer.fit(fold_train_feat)

            # Predictions (with probability) on train + val
            pred_train_full = mlp_model.transform(fold_train_feat)
            pred_val_full   = mlp_model.transform(fold_val_feat)

            # label index -> class name mapping from label indexer
            label_names = feat_model.stages[0].labels  # ['not_delayed','delayed'] (order)
            if "delayed" not in label_names:
                raise ValueError(f"'delayed' not found in labels: {label_names}")
            delayed_idx = float(label_names.index("delayed"))
            # binary case: the "other" class is the remaining index
            other_idx   = 1.0 - delayed_idx
            delayed_pos = int(delayed_idx)
            other_pos   = int(other_idx)

            # --- (3) Evaluate all thresholds for this fold ---
            for thr in threshold_list:
                train_p, train_r, train_f1 = prf_for_delayed_with_threshold(
                    pred_train_full, delayed_idx, other_idx, thr
                )
                val_p,   val_r,   val_f1   = prf_for_delayed_with_threshold(
                    pred_val_full, delayed_idx, other_idx, thr
                )

                metrics_acc[thr]["train_p"].append(train_p)
                metrics_acc[thr]["train_r"].append(train_r)
                metrics_acc[thr]["train_f1"].append(train_f1)
                metrics_acc[thr]["val_p"].append(val_p)
                metrics_acc[thr]["val_r"].append(val_r)
                metrics_acc[thr]["val_f1"].append(val_f1)

                print(
                    f"    Fold {fold_idx}, thr={thr:.2f} | "
                    f"TRAIN P={train_p:.3f}, R={train_r:.3f}, F1={train_f1:.3f} ; "
                    f"VAL P={val_p:.3f}, R={val_r:.3f}, F1={val_f1:.3f}"
                )
                # Confusion matrix on VALID for this fold+threshold (inline) 
                # 1) Extract P(delayed)
                df2 = pred_val_full.withColumn(
                    "prob_delayed",
                    vector_to_array("probability")[delayed_pos]
                )
                # 2) Threshold into custom prediction
                df_thr = df2.withColumn(
                    "pred_thr",
                    F.when(
                        F.col("prob_delayed") >= F.lit(thr),
                        F.lit(float(delayed_pos))
                    ).otherwise(F.lit(float(other_pos)))
                )
                # 3) Build confusion matrix as pandas DataFrame
                cm_pd = (
                    df_thr.groupBy("label", "pred_thr")
                          .count()
                          .toPandas()
                )
                if cm_pd.empty:
                    cm = pd.DataFrame(
                        0,
                        index=label_names,
                        columns=label_names
                    )
                else:
                    cm = (
                        cm_pd
                        .pivot(index="label", columns="pred_thr", values="count")
                        .fillna(0)
                        .astype(int)
                    )

                    # ensure full square matrix over all label indices
                    all_idx = list(range(len(label_names)))
                    cm = cm.reindex(index=all_idx, columns=all_idx, fill_value=0)

                    idx_to_name = {i: name for i, name in enumerate(label_names)}
                    cm.index   = [idx_to_name[i] for i in cm.index]
                    cm.columns = [idx_to_name[j] for j in cm.columns]
                # 4) Plot heatmap
                fig, ax = plt.subplots(figsize=(5, 4))
                sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", ax=ax)
                ax.set_xlabel("Predicted label")
                ax.set_ylabel("True label")
                ax.set_title(f"VALID Fold {fold_idx}  threshold={thr:.2f}")
                plt.tight_layout()
                try:
                    display(fig)  # Databricks
                except NameError:
                    plt.show()

                print(f"Confusion matrix (VALID) for fold {fold_idx}, thr={thr:.2f} plotted above.\n")
            # unpersist this fold's train features
            fold_train_feat.unpersist()

        # --- (4) Aggregate over folds and record per threshold ---
        for thr in threshold_list:
            tp = metrics_acc[thr]["train_p"]
            tr = metrics_acc[thr]["train_r"]
            tf = metrics_acc[thr]["train_f1"]
            vp = metrics_acc[thr]["val_p"]
            vr = metrics_acc[thr]["val_r"]
            vf = metrics_acc[thr]["val_f1"]

            mean_train_p = sum(tp) / len(tp)
            mean_train_r = sum(tr) / len(tr)
            mean_train_f = sum(tf) / len(tf)
            mean_val_p   = sum(vp) / len(vp)
            mean_val_r   = sum(vr) / len(vr)
            mean_val_f   = sum(vf) / len(vf)

            cfg_row = {
                "hidden_layers": str(hidden_layers),
                "maxIter": maxIter,
                "blockSize": blockSize,
                "threshold": thr,
                "train_precision_delayed_mean": mean_train_p,
                "train_recall_delayed_mean":   mean_train_r,
                "train_f1_delayed_mean":       mean_train_f,
                "val_precision_delayed_mean":  mean_val_p,
                "val_recall_delayed_mean":     mean_val_r,
                "val_f1_delayed_mean":         mean_val_f,
            }
            results.append(cfg_row)

            # track best by validation recall
            if mean_val_r > best_recall:
                best_recall = mean_val_r
                best_cfg = cfg_row

    # --- (5) Convert to table and sort by val recall ---
    results_df = pd.DataFrame(results)
    results_df_sorted = results_df.sort_values(
        by="val_recall_delayed_mean",
        ascending=False
    )

    print("\n===== BEST CONFIG (by mean validation recall for 'delayed') =====")
    print(best_cfg)

    print("\n===== GRID SEARCH RESULTS (sorted by val recall) =====")
    display(results_df_sorted)

    return results, best_cfg, results_df_sorted
```
